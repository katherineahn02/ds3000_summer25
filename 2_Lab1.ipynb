# Global Health
![alt text](GlobalHealth.jpg "Global Health")
## What is Global Health? 
Global health places a priority on achieving equity in health, emphasizing proactivity and finding solutions. As opposed to public health, global health is on a broader scale not only looking at a distinct community, but reaching beyond borders. This can be in following infectious diseases between countries and how it affects different populations around the world. Further, global health is multidisciplinary in that it considers various health determinants including but not limited to environmental factors and politics.
### How Politics Influence Global Health
The politics of a given country dictate funding and policy decisions for the health of its constituents. Global governments often set goals, enforce laws, contribute resources, and provide services in order to make decisions for care. Understandably, depending on factors beyond the scope of politics -- such as environmental -- there are different policies in place for different countries regarding health; this may also relate to the leading causes of death and the category it falls into such as communicable or noncommunicable disease or injuries. Below is a table of the leading cause of death in 2021 for each of four income groups based on gross national income, as well as the category of the leading cause of death.

| Income Group  | Leading Cause of Death      | Category
| ------------- | --------------------------- | --------
| Low           | Lower Respiratory Infection | Communicable Disease
| Lower-Middle  | COVID-19                    | Communicable Disease
| Upper-Middle  | Stroke                      | Noncommunicable Disease
| High          | Ischaemic Heart Disease     | Noncommunicable Disease

### Environmental Factors of Global Health
There are various examples of global environmental health issues, or rather how environmental factors may affect the health of a given population, with one of significance currently being climate change. This is as climate change 'threatens essential ingredients of good health.' According to the World Health Organization (WHO), the essential ingredients are 
- clean air
- safe drinking water
- nutritious food supply
- safe shelter'

You can read more from the WHO at [The Global Health Observatory](https://www.who.int/data/gho/data/themes/public-health-and-environment#:~:text=Climate%20change%20is%20impacting%20human,of%20progress%20in%20global%20health.)
